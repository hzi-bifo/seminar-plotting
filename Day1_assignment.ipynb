## Task 3: Errorbar

Create using AI tools

```python
df = pd.read_excel('mmc2.xlsx')
from scipy import stats
import numpy as np
# Display data structure
print("Data shape:", df.shape)
print("\nFirst few rows:")
#print(df.head())
print("\nUnique groups:", df['Group'].unique())
print("Unique timepoints:", sorted(df['timepoint after AB application'].unique()))

# ============================================================================
# 1. TIME-SERIES LINE PLOT WITH ERROR BARS
# ============================================================================

def plot_timeseries(df, strain='KB1', figsize=(12, 6)):
    """
    Create time-series plot showing bacterial abundance over time by group

    Parameters:
    -----------
    df : DataFrame
        Input data
    strain : str
        Bacterial strain to plot (e.g., 'KB1', 'YL2', etc.)
    """
    log_col = f'log10 {strain}'

    # Calculate mean and SEM for each group and timepoint
    summary = df.groupby(['Group', 'timepoint after AB application'])[log_col].agg([
        ('mean', 'mean'),
        ('sem', stats.std),  # Standard Error of Mean
        ('count', 'count')
    ]).reset_index()

    # Create plot
    fig, ax = plt.subplots(figsize=figsize)

    # Plot each group
    groups = summary['Group'].unique()
    colors = sns.color_palette("husl", n_colors=len(groups))

    for i, group in enumerate(groups):
        group_data = summary[summary['Group'] == group]

        ax.errorbar(
            group_data['timepoint after AB application'],
            group_data['mean'],
            yerr=group_data['sem'],
            marker='o',
            markersize=8,
            linewidth=2,
            capsize=5,
            capthick=2,
            label=group,
            color=colors[i],
            alpha=0.8
        )
    # Formatting
    ax.set_xlabel('Time after Antibiotic Application', fontsize=12)
    ax.set_ylabel(f'logâ‚â‚€ ({strain})', fontsize=12)
    ax.set_title(f'Bacterial Abundance Over Time - {strain}', fontsize=14)
    ax.legend(title='Group', frameon=True, loc='best')
    ax.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig(f'timeseries_{strain}.png', dpi=300, bbox_inches='tight')
    print(f"Saved: timeseries_{strain}.png")

    return fig, ax
plot_timeseries(df)
```

## a) Boxplot

```python
def plot_box_by_timepoint(df, strain='KB1', timepoints=None, figsize=(12, 6)):
    """
    Violin plot showing distribution shape of bacterial abundance
    Shows density of data at different values

    Parameters:
    -----------
    df : DataFrame
        Input data
    strain : str
        Bacterial strain to plot
    timepoints : list
        Specific timepoints to include
    """
    log_col = f'log10 {strain}'

    # Filter data
    if timepoints is not None:
        df_filtered = df[df['timepoint after AB application'].isin(timepoints)].copy()
    else:
        df_filtered = df.copy()

    df_filtered = df_filtered.dropna(subset=[log_col])

    # Create plot
    fig, ax = plt.subplots(figsize=figsize)

    # Violin plot
    sns.boxplot(
        data=df,
        x='timepoint after AB application',
        y=log_col,
        hue='Group',
        ax=ax,
        palette='husl',
        #inner='box',  # Shows box plot inside violin ('box', 'quartile', or None)
        linewidth=1.5,
    )

    # Overlay individual points
    sns.stripplot(
        data=df,
        x='timepoint after AB application',
        y=log_col,
        hue='Group',
        ax=ax,
        dodge=True,
        alpha=0.6,
        size=4,
        color='black',
        legend=False
    )

    # Formatting
    ax.set_xlabel('Time after Antibiotic Application', fontsize=14)
    ax.set_ylabel(f'logâ‚â‚€ ({strain})', fontsize=14)
    ax.set_title(f'Distribution of Bacterial Abundance (Violin Plot) - {strain}', fontsize=14)
    ax.legend(title='Group', frameon=True, loc='best')
    ax.grid(True, alpha=0.3, axis='y')

    plt.tight_layout()
    plt.savefig(f'violin_{strain}.png', dpi=300, bbox_inches='tight')
    print(f"Saved: violin_{strain}.png")

    return fig, ax
plot_box_by_timepoint(data)
```

```python
def plot_violin_by_timepoint(df, strain='KB1', timepoints=None, figsize=(12, 6)):
    """
    Violin plot showing distribution shape of bacterial abundance
    Shows density of data at different values

    Parameters:
    -----------
    df : DataFrame
        Input data
    strain : str
        Bacterial strain to plot
    timepoints : list
        Specific timepoints to include
    """
    log_col = f'log10 {strain}'

    # Filter data
    if timepoints is not None:
        df_filtered = df[df['timepoint after AB application'].isin(timepoints)].copy()
    else:
        df_filtered = df.copy()

    df_filtered = df_filtered.dropna(subset=[log_col])

    # Create plot
    fig, ax = plt.subplots(figsize=figsize)

    # Violin plot
    sns.violinplot(
        data=df,
        x='timepoint after AB application',
        y=log_col,
        hue='Group',
        ax=ax,
        palette='husl',
        inner='box',  # Shows box plot inside violin ('box', 'quartile', or None)
        linewidth=1.5,
        cut=0  # Limits violin to range of data
    )

    # Overlay individual points
    sns.stripplot(
        data=df,
        x='timepoint after AB application',
        y=log_col,
        hue='Group',
        ax=ax,
        dodge=True,
        alpha=0.6,
        size=4,
        color='black',
        legend=False
    )

    # Formatting
    ax.set_xlabel('Time after Antibiotic Application', fontsize=14)
    ax.set_ylabel(f'logâ‚â‚€ ({strain})', fontsize=14)
    ax.set_title(f'Distribution of Bacterial Abundance (Violin Plot) - {strain}', fontsize=14)
    ax.legend(title='Group', frameon=True, loc='best')
    ax.grid(True, alpha=0.3, axis='y')

    plt.tight_layout()
    plt.savefig(f'violin_{strain}.png', dpi=300, bbox_inches='tight')
    print(f"Saved: violin_{strain}.png")

    return fig, ax
plot_violin_by_timepoint(data)
```

## Task 1
Generate a histogram using the values

```python
def plot_grouped_bars(df, strain='KB1', timepoints=None, figsize=(10, 6)):
    """
    Create grouped bar plot comparing groups at specific timepoints

    Parameters:
    -----------
    df : DataFrame
        Input data
    strain : str
        Bacterial strain to plot
    timepoints : list
        List of timepoints to include (if None, uses all)
    """
    log_col = f'log10 {strain}'

    # Filter timepoints if specified
    if timepoints is not None:
        df_filtered = df[df['timepoint after AB application'].isin(timepoints)].copy()
    else:
        df_filtered = df.copy()

    # Calculate mean and SEM
    summary = df_filtered.groupby(['Group', 'timepoint after AB application'])[log_col].agg([
        ('mean', 'mean'),
        ('sem', stats.sem),
        ('count', 'count')
    ]).reset_index()

    # Create plot
    fig, ax = plt.subplots(figsize=figsize)

    # Set up bar positions
    groups = summary['Group'].unique()
    timepoints_plot = sorted(summary['timepoint after AB application'].unique())
    x = np.arange(len(timepoints_plot))
    width = 0.8 / len(groups)  # Bar width

    colors = sns.color_palette("husl", n_colors=len(groups))

    # Plot bars for each group
    for i, group in enumerate(groups):
        group_data = summary[summary['Group'] == group].sort_values('timepoint after AB application')

        positions = x + (i - len(groups)/2 + 0.5) * width

        ax.bar(
            positions,
            group_data['mean'],
            width,
            yerr=group_data['sem'],
            label=group,
            color=colors[i],
            alpha=0.8,
            capsize=5,
            error_kw={'linewidth': 2, 'elinewidth': 2}
        )

    # Formatting
    ax.set_xlabel('Time after Antibiotic Application', fontsize=12, fontweight='bold')
    ax.set_ylabel(f'logâ‚â‚€ CFU/g ({strain})', fontsize=12, fontweight='bold')
    ax.set_title(f'Bacterial Abundance Comparison - {strain}', fontsize=14, fontweight='bold')
    ax.set_xticks(x)
    ax.set_xticklabels(timepoints_plot)
    ax.legend(title='Group', frameon=True)
    ax.grid(True, alpha=0.3, axis='y')

    plt.tight_layout()
    plt.savefig(f'grouped_bars_{strain}.png', dpi=300, bbox_inches='tight')
    print(f"Saved: grouped_bars_{strain}.png")

    return fig, ax
plot_grouped_bars(df)
```
